```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
df = pd.read_csv('bank_region_r.csv', header=0, encoding='utf-8')
bank,region,bsum,cover_r,label = df.ix[0:,1:2],df.ix[0:,2:3],df.ix[0:,3:4],df.ix[0:,4:5],df.ix[0:,5:6]
plt.plot(bank,cover_r,'r*',label='bank--r')
plt.legend(loc='upper right')
plt.show()
plt.plot(region,cover_r,'b*',label='region--r')
plt.legend(loc='upper right')
plt.show()
plt.plot(label,cover_r,'g*',label='label--r')
plt.legend(loc='upper right')
plt.show()

plt.plot(bank,bsum,'r*',label='bank--bsum')
plt.legend(loc='upper right')
plt.show()
plt.plot(region,bsum,'b*',label='region--bsum')
plt.legend(loc='upper right')
plt.show()
plt.plot(label,bsum,'g*',label='label--bsum')
plt.legend(loc='upper right')
plt.show()

# fig1 = plt.figure()
# ax = Axes3D(fig1)
# X, Y,Z = bank, region,cover_r
# ax.scatter(X,Y,Z,c='g')
# ax.set_zlabel('cover_r') #坐标轴
# ax.set_ylabel('region')
# ax.set_xlabel('bank')
# plt.show()
# fig2 = plt.figure()
# ax = Axes3D(fig2)
# X, Y,Z = bank, region,bsum
# ax.scatter(X,Y,Z,c='r')
# ax.set_zlabel('bsum') #坐标轴
# ax.set_ylabel('region')
# ax.set_xlabel('bank')
# plt.show()
```
